## Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
dataset = pd.read_csv("Social_Network_Ads.csv")
X = dataset.iloc[:, :-1].values
Y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

## Training the K-NN model on the Training set

```python
from sklearn.neighbors import KNeighborsClassifier
```

metric='minkowski', p = 2 ensures that it's EUCLIDIAN DISTANCE: which is basically PYTHAGORAS THEOREM in COORDINATE GEOMETRY

```python
classifier = KNeighborsClassifier(n_neighbors = 5, weights='uniform', #weights means all points are weighted equally
                                 metric='minkowski', p = 2)  #metric: method to calculate distance between observation point and neighbors
classifier.fit(X_train, Y_train)
```

## Predicting a new result

```python
classifier.predict(sc.transform([[30, 87000]]))
```

## Predicting the Test set results

```python
Y_pred = classifier.predict(X_test)
```

<b>PREDICTED RESULTS VS ACTUAL RESULTS</b>

```python
np.concatenate((Y_pred.reshape(-1,1), Y_test.reshape(-1,1)), axis = 1)
```

## Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(Y_test, Y_pred)
cm
```

```python
accuracy_score(Y_test, Y_pred)
```

```python
(cm[0][0] + cm[1][1]) / np.sum(cm)
```

<b>4%</b> BETTER THAN LOGISTIC REGRESSION

```python
sc.inverse_transform(X_train)[:,0].max()
```

## Visualising the Training set results

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), Y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 2),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 2))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

## Visualising the Test set results

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_test), Y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 2),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 2))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

THOUGHTS: Prediction is not a straight line now. In logistic regression, it was a straight line and therefore resulted in incorrect predictions. Now we have a flexible curve which does much better.
